```python
import math
import numpy as np
```

```python
def sin_value(angle):
    angle_in_degrees = angle
    angle_radians = math.radians(angle_in_degrees)
    return math.sin(angle_radians)
    
def cos_value(angle):
    return np.sqrt(1 - (sin_value(angle))**2)

def tan_value(angle):
    return sin_value(angle)/cos_value(angle)

def cosec_value(angle):
    return 1/sin_value(angle)

def sec_value(angle):
    return 1/cos_value(angle)

def cot_value(angle):
    return 1/tan_value(angle)
```

```python
import matplotlib.pyplot as plt

# Function to draw coordinate axes
def draw_axes():
    fig, ax = plt.subplots()    
    plt.scatter(x=4, y=3)
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis', rotation=0, ha='right')
    ax.set_title('Coordinate Axes')
    plt.show()

# Call the function to draw coordinate axes
draw_axes()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Function to draw rotated coordinate axes
def draw_rotated_axes(theta_degrees=30):
    # Convert angle to radians
    theta_radians = np.radians(theta_degrees)

    # Rotation matrix
    rotation_matrix = np.array([[np.cos(theta_radians), -np.sin(theta_radians)],
                                [np.sin(theta_radians), np.cos(theta_radians)]])

    # Original axes
    x_axis = np.array([1, 0])
    y_axis = np.array([0, 1])

    # Apply rotation to axes
    rotated_x_axis = np.dot(rotation_matrix, x_axis)
    rotated_y_axis = np.dot(rotation_matrix, y_axis)

    # Plot rotated axes
    fig, ax = plt.subplots()
    ax.quiver(0, 0, rotated_x_axis[0], rotated_x_axis[1], angles='xy',
              scale_units='xy', scale=1, color='red', label='Rotated X-axis')
    ax.quiver(0, 0, rotated_y_axis[0], rotated_y_axis[1], angles='xy',
              scale_units='xy', scale=1, color='blue', label='Rotated Y-axis')

    # Set labels and title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title(f'Rotated Axes by {theta_degrees} degrees')

    # Set axis limits
    ax.set_xlim([-1, 1])
    ax.set_ylim([-1, 1])

    # Add legend
    ax.legend()

    # Show the plot
    plt.grid()
    plt.show()

# Call the function to draw rotated coordinate axes
draw_rotated_axes(theta_degrees=90)
```
