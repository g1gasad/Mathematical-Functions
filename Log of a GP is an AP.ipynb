```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
def geometric_progression(first_term, common_ratio, number_of_terms, plot_log=False):
    a, r, n = first_term, common_ratio, number_of_terms
    series = [a*(r)**(x-1) for x in range(1, n+1)]
    if plot_log == True:
        log_series = np.log(series)
        plt.figure(figsize=(10,5))
        sns.lineplot(log_series, ls="--", marker="D", markerfacecolor='teal', markersize=6)
        plt.show()
    else:
        plt.figure(figsize=(10,5))
        sns.lineplot(series, ls="--", marker="D", markerfacecolor='teal', markersize=6)
        plt.show()
geometric_progression(100, -49/50, 100)
```
